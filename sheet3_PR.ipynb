```python
import numpy as np
import sklearn 
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import math
from sklearn.neighbors import kneighbors_graph

data=np.array([[5,8],[10,8],[11,8],[6,7],[10,7],[12,7],[13,7],[5,6],
                   [10,6],[13,6],[6,5],[9,4],[11,5],[14,6],[15,5],[2,4],
                   [3,4],[5,4],[6,4],[7,4],[15,4],[3,3],[7,3],[8,2]])

A=rbf_kernel(data,gamma=0.01)
print(np.shape(A))
degree = np.diag(np.sum(A, axis=1))
print(np.shape(degree))
B=np.dot(np.linalg.inv(degree) , (degree-A))
print(np.shape(B))
eigenVal,eigenVec=np.linalg.eigh(B)
idx = eigenVal.argsort()[::-1]
eigenVal = eigenVal[idx]
eigenVec = eigenVec[:,idx]
print(eigenVal)
print(np.shape(eigenVec))
norm_eig = eigenVec[21:]
norm_eig = norm_eig.T
print(np.shape(norm_eig))
for i in range(24):
  for j in range(3):
    norm_eig[i][j]=norm_eig[i][j]/math.sqrt(norm_eig[i][0]**2+norm_eig[i][1]**2+norm_eig[i][2]**2)
print(norm_eig)

x = norm_eig.T[0]
y= norm_eig.T[1]
z=norm_eig.T[2]
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, color = "green")
plt.title("normalized eigenVectors , gamma = 0.01")
plt.show()

A=rbf_kernel(data,gamma=0.1)
print(np.shape(A))
degree = np.diag(np.sum(A, axis=1))
print(np.shape(degree))
B=np.dot(np.linalg.inv(degree) , (degree-A))
print(np.shape(B))
eigenVal,eigenVec=np.linalg.eigh(B)
idx = eigenVal.argsort()[::-1]
eigenVal = eigenVal[idx]
eigenVec = eigenVec[:,idx]
print(eigenVal)
print(np.shape(eigenVec))
norm_eig = eigenVec[21:]
norm_eig = norm_eig.T
print(np.shape(norm_eig))
for i in range(24):
  for j in range(3):
    norm_eig[i][j]=norm_eig[i][j]/math.sqrt(norm_eig[i][0]**2+norm_eig[i][1]**2+norm_eig[i][2]**2)
print(norm_eig)

x = norm_eig.T[0]
y= norm_eig.T[1]
z=norm_eig.T[2]
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, color = "blue")
plt.title("normalized eigenVectors , gamma = 0.1")
plt.show()

A=rbf_kernel(data,gamma=1)
print(np.shape(A))
degree = np.diag(np.sum(A, axis=1))
print(np.shape(degree))
B=np.dot(np.linalg.inv(degree) , (degree-A))
print(np.shape(B))
eigenVal,eigenVec=np.linalg.eigh(B)
idx = eigenVal.argsort()[::-1]
eigenVal = eigenVal[idx]
eigenVec = eigenVec[:,idx]
print(eigenVal)
print(np.shape(eigenVec))
norm_eig = eigenVec[21:]
norm_eig = norm_eig.T
print(np.shape(norm_eig))
for i in range(24):
  for j in range(3):
    norm_eig[i][j]=norm_eig[i][j]/math.sqrt(norm_eig[i][0]**2+norm_eig[i][1]**2+norm_eig[i][2]**2)
print(norm_eig)

x = norm_eig.T[0]
y= norm_eig.T[1]
z=norm_eig.T[2]
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, color = "red")
plt.title("normalized eigenVectors , gamma = 1")
plt.show()

A=rbf_kernel(data,gamma=10)
print(np.shape(A))
degree = np.diag(np.sum(A, axis=1))
print(np.shape(degree))
B=np.dot(np.linalg.inv(degree) , (degree-A))
print(np.shape(B))
eigenVal,eigenVec=np.linalg.eigh(B)
idx = eigenVal.argsort()[::-1]
eigenVal = eigenVal[idx]
eigenVec = eigenVec[:,idx]
print(eigenVal)
print(np.shape(eigenVec))
norm_eig = eigenVec[21:]
norm_eig = norm_eig.T
print(np.shape(norm_eig))
for i in range(24):
  for j in range(3):
    norm_eig[i][j]=norm_eig[i][j]/math.sqrt(norm_eig[i][0]**2+norm_eig[i][1]**2+norm_eig[i][2]**2)
print(norm_eig)

x = norm_eig.T[0]
y= norm_eig.T[1]
z=norm_eig.T[2]
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, color = "yellow")
plt.title("normalized eigenVectors , gamma = 10")
plt.show()

A = kneighbors_graph(data,3,mode='connectivity').toarray()
print(np.shape(A))
degree = np.diag(np.sum(A, axis=1))
print(np.shape(degree))
B=np.dot(np.linalg.inv(degree) , (degree-A))
print(np.shape(B))
eigenVal,eigenVec=np.linalg.eigh(B)
idx = eigenVal.argsort()[::-1]
eigenVal = eigenVal[idx]
eigenVec = eigenVec[:,idx]
print(eigenVal)
print(np.shape(eigenVec))
norm_eig = eigenVec[21:]
norm_eig = norm_eig.T
print(np.shape(norm_eig))
for i in range(24):
  for j in range(3):
    norm_eig[i][j]=norm_eig[i][j]/math.sqrt(norm_eig[i][0]**2+norm_eig[i][1]**2+norm_eig[i][2]**2)
print(norm_eig)

x = norm_eig.T[0]
y= norm_eig.T[1]
z=norm_eig.T[2]
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, color = "magenta")
plt.title("normalized eigenvectors, 3-NN graph")
plt.show()


# here is clustering data using sickit.learn:
clustering = SpectralClustering(n_clusters=3, gamma = 0.01 , assign_labels='discretize', random_state=0).fit(data)
print(clustering.labels_)
clustering2 = SpectralClustering(n_clusters=3, gamma = 0.1 , assign_labels='discretize', random_state=0).fit(data)
print(clustering2.labels_)
clustering3 = SpectralClustering(n_clusters=3, gamma = 1 , assign_labels='discretize', random_state=0).fit(data)
print(clustering3.labels_)
clustering4 = SpectralClustering(n_clusters=3, gamma = 10 , assign_labels='discretize', random_state=0).fit(data)
print(clustering4.labels_)

colors = ['blue', 'red','green']
for label, color in zip(range(len(clustering.labels_)), colors):
  plt.scatter(data[clustering.labels_==label, 0],
              data[clustering.labels_==label, 1],
              label=np.unique(clustering.labels_)[label],
              alpha = 1, 
              color=color)
  plt.xlabel(data[0])
  plt.ylabel(data[1])
  plt.legend()
  plt.title('clustering with gamma = 0.01')
plt.show()

colors = ['blue', 'red','green']
for label, color in zip(range(len(clustering2.labels_)), colors):
  plt.scatter(data[clustering2.labels_==label, 0],
              data[clustering2.labels_==label, 1],
              label=np.unique(clustering2.labels_)[label],
              alpha = 1, 
              color=color)
  plt.xlabel(data[0])
  plt.ylabel(data[1])
  plt.legend()
  plt.title('clustering with gamma = 0.1')
plt.show()

colors = ['blue', 'red','green']
for label, color in zip(range(len(clustering3.labels_)), colors):
  plt.scatter(data[clustering3.labels_==label, 0],
              data[clustering3.labels_==label, 1],
              label=np.unique(clustering3.labels_)[label],
              alpha = 1, 
              color=color)
  plt.xlabel(data[0])
  plt.ylabel(data[1])
  plt.legend()
  plt.title('clustering with gamma = 1')
plt.show()
colors = ['blue', 'red','green']
for label, color in zip(range(len(clustering4.labels_)), colors):
  plt.scatter(data[clustering4.labels_==label, 0],
              data[clustering4.labels_==label, 1],
              label=np.unique(clustering4.labels_)[label],
              alpha = 1, 
              color=color)
  plt.xlabel(data[0])
  plt.ylabel(data[1])
  plt.legend()
  plt.title('clustering with gamma = 10')
plt.show()
```
