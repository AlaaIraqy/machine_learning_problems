```python
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import mpl_toolkits.mplot3d
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.linalg as la


#QUESTION 2:

data=np.array([[10, 60, 10, 90], [20, 50, 40, 70], [30, 50, 30, 40], [20, 50, 20, 60], [10, 60, 30, 10]])
norms= np.linalg.norm(data, axis = 1)
print("Norms of each instance:")
print(norms)
print("cosines:")
dist_out = 1-pairwise_distances(data, metric="cosine")
print(dist_out)
print("Distances")
eucDist=euclidean_distances(data, data)
print(eucDist)

#QUESTION 3:

df = pd.DataFrame([[10, 60,90], [20,50, 70],[30,50,40],[20,50,60],[10,60,10]], columns=["a1", "a2", "a3"])
print(df)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.T[0], df.T[1], df.T[2])
plt.show()
mean=np.mean(df, axis=0)
print ("Mean : ", mean)
centered = df-mean
print("centered data: " ,centered)
covMatrix = np.cov(df.T,bias=True)
print ("Cov matrix=",covMatrix)
w, v = la.eig(covMatrix)
wdiag=np.diag(w)
print("eigen values: ",wdiag)
print("eigen vectors: ",v)
res=np.dot(np.dot(v,wdiag),v.T)
print("Check:")
print(res)
rdiag=np.flip(wdiag)
print(rdiag )
rv=np.flip(v,axis=1)
print(rv)
nr=rv[: ,  0:2]
print(nr)
p=np.dot(nr.T, centered.T).T
print(p)
x1=p[:, 0]
x2=p[:, 1]
plt.scatter(x1, x2)
plt.show()
```
