```python
import mpl_toolkits.mplot3d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
data = load_iris()

#QUESTION_1:

plt.figure(figsize=(5, 4))
plt.hist(data.data.T[0], color='red', bins=20)
plt.title(data.feature_names[0])
plt.xlabel(data.feature_names[0])
plt.ylabel("count")

plt.figure(figsize=(5, 4))
plt.hist(data.data.T[1], bins=20, color='red')
plt.title(data.feature_names[1])
plt.xlabel(data.feature_names[1])
plt.ylabel("count")

plt.figure(figsize=(5, 4))
plt.hist(data.data.T[2], bins=20, color='red')
plt.title(data.feature_names[2])
plt.xlabel(data.feature_names[2])
plt.ylabel("count")

plt.figure(figsize=(5, 4))
plt.hist(data.data.T[3], bins=20, color='red')
plt.title(data.feature_names[3])
plt.xlabel(data.feature_names[3])
plt.ylabel("count")
plt.show()

#QUESTION_2:

formatter = plt.FuncFormatter(lambda i, *args: data.target_names[int(i)])
plt.figure(figsize=(5, 4))
plt.scatter(data.data[:, 0], data.data[:, 1], c=data.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])

plt.figure(figsize=(5, 4))
plt.scatter(data.data[:, 0], data.data[:, 2], c=data.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[2])

plt.figure(figsize=(5, 4))
plt.scatter(data.data[:, 0], data.data[:, 3], c=data.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[3])

plt.figure(figsize=(5, 4))
plt.scatter(data.data[:, 1], data.data[:, 2], c=data.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(data.feature_names[1])
plt.ylabel(data.feature_names[2])

plt.figure(figsize=(5, 4))
plt.scatter(data.data[:, 1], data.data[:, 3], c=data.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(data.feature_names[1])
plt.ylabel(data.feature_names[3])

plt.figure(figsize=(5, 4))
plt.scatter(data.data[:, 2], data.data[:, 3], c=data.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(data.feature_names[2])
plt.ylabel(data.feature_names[3])
plt.tight_layout()
plt.show()



# #QUESTION_3:

fig1=plt.figure()
g1 = fig1.add_subplot(111, projection="3d")
g1.scatter(data.data[:, 0], data.data[:, 1], data.data[:, 2], c=data.target)
g1.set_xlabel(data.feature_names[0])
g1.set_ylabel(data.feature_names[1])
g1.set_zlabel(data.feature_names[2])

fig2=plt.figure()
g1 = fig2.add_subplot(111, projection="3d")
g1.scatter(data.data[:, 0], data.data[:, 1], data.data[:, 3], c=data.target)
g1.set_xlabel(data.feature_names[0])
g1.set_ylabel(data.feature_names[1])
g1.set_zlabel(data.feature_names[3])

fig3=plt.figure()
g1 = fig3.add_subplot(111, projection="3d")
g1.scatter(data.data[:, 0], data.data[:, 2], data.data[:, 3], c=data.target)
g1.set_xlabel(data.feature_names[0])
g1.set_ylabel(data.feature_names[2])
g1.set_zlabel(data.feature_names[3])

fig4=plt.figure()
g1 = fig4.add_subplot(111, projection="3d")
g1.scatter(data.data[:, 1], data.data[:, 2], data.data[:, 3], c=data.target)
g1.set_xlabel(data.feature_names[1])
g1.set_ylabel(data.feature_names[2])
g1.set_zlabel(data.feature_names[3])


plt.tight_layout()
plt.show()

#Question_4:

df = pd.DataFrame(data.data, columns=data.feature_names)
print(df.corr())
# another way
matrix = np.corrcoef(data.data.T)
print(matrix)

#QUESTION_5:

plt.figure()
df.boxplot()
plt.show()
```
